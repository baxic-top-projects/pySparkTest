```python
import requests
MLFLOW_SERVE = "http://mlflow-serve:5001"  # или localhost:5001 с хоста
def predict_spell(mana_cost: float, damage: float) -> int:
    r = requests.post(
        f"{MLFLOW_SERVE}/invocations",
        json={
            "dataframe_split": {
                "columns": ["mana_cost", "damage"],
                "data": [[mana_cost, damage]],
            }
        },
        timeout=10,
    )
    r.raise_for_status()
    return r.json()["predictions"][0]
```

```python
predict_spell(5,4)
```
